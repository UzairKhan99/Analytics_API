### Get All NoteBook Method 


```python
import requests

path = "/api/events"
base_url = "http://127.0.0.1:8002"
endpoint = f"{base_url}{path}"

response = requests.get(endpoint)

if response.ok:
    data = response.json()
    print(data)
    print(data.get("results"))
    print(data.get("count"))
else:
    print("error", response.status_code, response.text)
```

### Post method 


```python
import requests

path = "/api/events"
base_url = "http://127.0.0.1:8002"
endpoint = f"{base_url}{path}"

payload = {"page": 13}

response = requests.post(endpoint, json=payload)

if response.ok:
    data = response.json()
    print(data)
    print(data.get("page"))
else:
    print("error", response.status_code, response.text)
```

### Put Method 

```python
import requests

path = "/api/events/12"
base_url = "http://127.0.0.1:8002"
endpoint = f"{base_url}{path}"

payload = {"description": "Hello how are you"}

response = requests.put(endpoint, json=payload)

if response.ok:
    data = response.json()
    print(data)
    print(data.get("id"))
else:
    print("error", response.status_code, response.text)
```
